```python
import json
import nltk
import random
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
with open('/content/intents (1).json', 'r') as f:
    intents = json.load(f)
```

```python
nltk.download('punkt') # first-time use only
nltk.download('wordnet') # first-time use only
```

```python
from nltk.classify import scikitlearn
# Tokenize the patterns
tokenizer = nltk.RegexpTokenizer(r'\w+')
for intent in intents['intents']:
    intent['tokenized_patterns'] = [tokenizer.tokenize(pattern.lower()) for pattern in intent['patterns']]
    
# Lemmatize the words
lemmatizer = nltk.WordNetLemmatizer()
for intent in intents['intents']:
    intent['lemmatized_patterns'] = [[lemmatizer.lemmatize(word) for word in pattern] for pattern in intent['tokenized_patterns']]

# Create a bag of words representation of the patterns
vectorizer = CountVectorizer(tokenizer=lambda doc: doc, lowercase=False)
X = [word for intent in intents['intents'] for word in intent['lemmatized_patterns']]
y = np.array([intents['intents'] for intent in intents['intents']])
```
